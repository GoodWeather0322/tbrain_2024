```python
from opencc import OpenCC
import re

opencc = OpenCC("s2t")


def _opencc_convert(text: str):
    text = opencc.convert(text)
    return text

def _remove_stopwords(text: str):
    text = re.sub(r"\*\*page \d+\*\*", "", text)
    text = re.sub(r"\*\*question \d+\*\*", "", text)
    text = re.sub(r"\*\*answer \d+\*\*", "", text)
    # Step 1: 去除網址和 EMAIL
    text = re.sub(r"http\S+|www\S+|https\S+|[\w\.-]+@[\w\.-]+", "", text)

    text = re.sub(r"【[A-Za-z0-9]+】", "", text)

    # Step 6: 去除 "第 X 頁，共 Y 頁" 格式
    text = re.sub(r"第 \d+ 頁，共 \d+ 頁", "", text)

    # Step 7: 去除 "X/Y" 或 "X / Y" 格式
    text = re.sub(r"\b\d+ ?/ ?\d+\b", "", text)

    # Step 8: 去除 "~X~" 格式
    text = re.sub(r"~\d+~", "", text)

    # Step 9: 去除 "（接次頁）" 和 "（承前頁）"
    text = re.sub(r"（接次頁）|（承前頁）", "", text)

    # Step 10: 去除 "- X -" 格式
    text = re.sub(r"- \d+ -", "", text)

    # Step 2: 去除無意義數字（可以依需求調整，如果想保留某些數字格式）
    text = re.sub(r"\b\d+\b", "", text)

    # Step 3: 去除標點符號
    text = re.sub(r"[^\w\s]", "", text)

    # 去除多餘的空格
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
max_tokens = 128
stride = 64

question_file = "/Users/goodweather/Desktop/workspace/tbrain_2024/source/競賽資料集/dataset/preliminary/questions_example.json"

with open(question_file, "r") as f:
    questions = json.load(f)['questions']

for question in questions:
    source = question['source']
    category = question['category']
    for idx in source:
        # if category != "finance" or idx != 803:
        #     continue
        text_file = f'/Users/goodweather/Desktop/workspace/tbrain_2024/source/競賽資料集/reference_text/{category}/{idx}.txt'
        with open(text_file, 'r') as f:
            text = f.read()
            text = _opencc_convert(text)
            text = _remove_stopwords(text)

            tokens_list = embedding_model.tokenizer.encode(text, add_special_tokens=False, max_length=max_tokens, stride=stride, return_overflowing_tokens=True)

            split_texts = embedding_model.tokenizer.batch_decode(tokens_list)

            for i, split_text in enumerate(split_texts):
                tokens = embedding_model.tokenizer.encode(split_text, add_special_tokens=False)
                if i == len(split_texts) - 1:
                    pass
                else:
                    if len(tokens) == 128:
                        pass
                    else:
                        text = embedding_model.tokenizer.decode(tokens, skip_special_tokens=True)
                        print(category, idx, i)
                        # print(text)
                        print(len(embedding_model.tokenizer.encode(text, add_special_tokens=False)))

```
